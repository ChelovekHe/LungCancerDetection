```python
from paths import *
from sklearn.cluster import KMeans
from glob import glob
from tqdm import tqdm
from os.path import join, isdir, basename
%pylab inline
```

```python
nodules = array(glob(join(PATH['LUNA_NODULES'], 'subset*', '*.npy')))
vessels = array(glob(join(PATH['LUNA_VESSELS'], 'subset*', '*.npy')))
```

```python
# similars = list()
for nodule_path in nodules[len(similars):]:
    mindist = 1e+20
    for vessel_path in tqdm(choice(vessels, 100000, False)):
        nodule = load(nodule_path)
        vessel = load(vessel_path)
        dist = linalg.norm(nodule[10: 54,
                                  10: 54, 
                                  10: 54].flatten() 
                           - vessel[10: 54,
                                    10: 54, 
                                    10: 54].flatten(), 2)
        if dist < mindist:
            mindist = dist
            similar = (nodule_path, vessel_path)
    
    similars.append(similar)
```
